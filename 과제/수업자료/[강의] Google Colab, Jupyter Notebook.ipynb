## Google Colab 특징
- 구글 클라우드 기반의 무료 개발 환경 서비스

- 여러 사용자와 동시에 사용 가능
- PC, 태블릿, 모바일 상관없이 인터넷 브라우저만 있으면 언제 어디서나 접속 가능

- 구글 드라이브와 연동이 가능
- Git이나 Github와 쉽게 연동 가능

## 매직 명령어(magic commands)

* Jupyter Notebook 환경에서 파이썬 파일 저장 및 실행

```
%%writefile test.py
print('Hello Colab')
```

```
%run test.py
```

* 코드 실행 시간 측정

```
%%timeit?
```

```
%timeit li = [n ** n for n in range(1000)]
```

```
%%timeit
li = []
for n in range(10000):
  li.append(n ** n)
```

## 셸 명령어

* 텍스트 기반의 셸 명령어 처리
* `!` 문자를 명령어 앞에 붙여서 셸 명령어 사용 가능

* `ls`: 디렉토리 리스트 보기

```
!ls
```

* `pwd`: 현재 경로 보기

```
!pwd
```

* `cd`: 디렉토리 변경
* IPython에서는 임시 셸에서 실행

```
!cd sample_data/ && ls
```

* `%cd`: 지속적인 디렉토리 변경

```
%cd sample_data
```

* `echo`: 화면 출력

* `mkdir`: 디렉토리 생성

* `cat`: 파일 보기

* `cp`: 디렉토리/파일 복사

* `rm`: 디렉토리/파일 삭제
